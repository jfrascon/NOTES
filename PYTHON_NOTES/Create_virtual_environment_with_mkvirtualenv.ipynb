# Create virtual environment for Python2 and Python3

## Install `virtualenv` and `virtualenvwrapper` packages

For Python3  

```bash
$ pip3 install virtualenv virtualenvwrapper
$ echo "source /usr/local/bin/virtualenvwrapper.sh" >> ~/.bash_profile
$ source ~/.bash_profile
```

For Python2  

```bash
$ pip2 install virtualenv virtualenvwrapper
$ echo "source /usr/local/bin/virtualenvwrapper.sh" >> ~/.bash_profile
$ source ~/.bash_profile
```

## Create virtual environment for Python3 and use it.

```bash
# Create virual environment
$ mkvirtualenv <envname> -p python3

# Activate virtual environment
$ workon <envname>

# Install python packages/modules in the virtual environment
$ pip3 install numpy scipy matplotlib scikit-image scikit-learn ipython

# Deactivate virtual environment
$ deactivate
```

Link additional libraries to the virual environment if needed

```bash
$ cd ~/.virtualenvs/envname/lib/python3.6/site-packages/
$ ln -s /usr/local/Cellar/opencv/3.3.1_1/lib/python3.6/site-packages/cv2.cpython-36m-darwin.so cv2.so
```

## Create virtual environment for Python2 and use it.

```bash
# Create virual environment
$ mkvirtualenv <envname> -p python2

# Activate virtual environment
$ workon <envname>

# Install python packages/modules in the virtual environment
$ pip2 install numpy scipy matplotlib scikit-image scikit-learn ipython

# Deactivate virtual environment
$ deactivate
```

Link additional libraries to the virual environment if needed

```bash
$ cd ~/.virtualenvs/envname/lib/python2.7/site-packages/
$ ln -s /usr/local/Cellar/opencv/3.3.1_1/lib/python2.7/site-packages/cv2.so cv2.so
```
